```python
import pandas as pd

# Part 1: Data Loading and Exploration
# Read all CSV files into Pandas DataFrames

products = pd.read_csv('products.csv')
sales = pd.read_csv('sales.csv')
inventory = pd.read_csv('inventory.csv')
customers = pd.read_csv('customers1.csv')
stores = pd.read_csv('stores.csv')

# Display the first few rows of each DataFrame
print("\n=== products ===")
print(products.head(5))
print("\n=== sales ===")
print(sales.head(5))
print("\n=== inventory ===")
print(inventory.head(5))
print("\n=== customers ===")
print(customers.head(5))
print("\n=== stores ===")
print(stores.head(5))
```

```python
# 2. Exploratory Analysis

# Obtain summary statistics for numerical columns

print("\n=== Info products ===")
print(products.describe())
print("\n=== Info sales ===")
print(sales.describe())
print("\n=== Info inventory ===")
print(inventory.describe())

# Check for missing values in all dataframes
print("\n=== Missing Values products ===")
print(products.isnull().sum())
print("\n=== Missing Values sales ===")
print(sales.isnull().sum())
print("\n=== Missing Values inventory ===")
print(inventory.isnull().sum())
print("\n=== Missing Values customers ===")
print(customers.isnull().sum())
print("\n=== Missing Values stores ===")
print(stores.isnull().sum())

# Display the data types of each column
print("\n=== Data Types products ===")
print(products.dtypes)
print("\n=== Data Types sales ===")
print(sales.dtypes)
print("\n=== Data Types inventory ===")
print(inventory.dtypes)
print("\n=== Data Types customers ===")
print(customers.dtypes)
print("\n=== Data Types stores ===")
print(stores.dtypes)
```

```python
# Part 2: Data Cleaning

# 1. Handling Missing Values

# Create copies of original DataFrames for cleaning

print("\n=== products ===")
print(products.isnull().sum())
print("\n=== sales ===")
print(sales.isnull().sum())
print("\n=== inventory ===")
print(inventory.isnull().sum())
print("\n=== customers ===")
print(customers.isnull().sum())
print("\n=== stores ===")
print(stores.isnull().sum())

# For numerical columns with missing values, replace with column mean

numerical_columns  = products.select_dtypes(include=['number']).columns
products[numerical_columns] = products[numerical_columns ].fillna(products[numerical_columns].mean())
numerical_columns  = sales.select_dtypes(include=['number']).columns
sales[numerical_columns] = sales[numerical_columns ].fillna(sales[numerical_columns].mean())
numerical_columns  = inventory.select_dtypes(include=['number']).columns
inventory[numerical_columns] = inventory[numerical_columns ].fillna(inventory[numerical_columns].mean())
numerical_columns  = customers.select_dtypes(include=['number']).columns
customers[numerical_columns] = customers[numerical_columns ].fillna(customers[numerical_columns].mean())
numerical_columns  = stores.select_dtypes(include=['number']).columns
stores[numerical_columns] = stores[numerical_columns ].fillna(stores[numerical_columns].mean())

# For categorical columns with missing values, replace with most frequent value

categorical_columns  = products.select_dtypes(include=['object', 'category']).columns
for column in categorical_columns :
    products[column] = products[column].fillna(products[column].mode()[0])
categorical_columns  = sales.select_dtypes(include=['object', 'category']).columns
for column in categorical_columns :
    sales[column] = sales[column].fillna(sales[column].mode()[0])       
categorical_columns  = inventory.select_dtypes(include=['object', 'category']).columns
for column in categorical_columns :
    inventory[column] = inventory[column].fillna(inventory[column].mode()[0])       
categorical_columns  = customers.select_dtypes(include=['object', 'category']).columns
for column in categorical_columns :
    customers[column] = customers[column].fillna(customers[column].mode()[0])    
categorical_columns  =  stores.select_dtypes(include=['object', 'category']).columns
for column in categorical_columns :
    stores[column] = stores[column].fillna(stores[column].mode()[0])      

# For date columns with missing values, use forward fill or backward fill 

date_columns  = products.select_dtypes(include=['datetime']).columns
products[date_columns] = products[date_columns].ffill().bfill()
date_columns  = sales.select_dtypes(include=['datetime']).columns
sales[date_columns] = sales[date_columns].ffill().bfill()
date_columns  = inventory.select_dtypes(include=['datetime']).columns
inventory[date_columns] = inventory[date_columns].ffill().bfill()
date_columns  = customers.select_dtypes(include=['datetime']).columns
customers[date_columns] = customers[date_columns].ffill().bfill()
date_columns  =  stores.select_dtypes(include=['datetime']).columns
stores[date_columns] = stores[date_columns].ffill().bfill()
```

```python
# 2. Removing Duplicates:

print("\n=== Duplicates ===")
print(customers.duplicated().sum()) # to identified rows where all column values match a previous row

products = products.drop_duplicates() 
customers = customers.drop_duplicates()# to ensure only unique entries remain
```
